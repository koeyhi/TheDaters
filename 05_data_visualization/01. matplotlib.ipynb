# 데이터 시각화
- 데이터 시각화는 복잡하고 방대한 양의 데이터를 시각적인 요소를 활용하여 쉽고 빠르게 이해할 수 있도록 표현하는 과정
- 다양한 시각적 도구를 사용하여 데이터의 특징, 추세, 패턴을 한눈에 파악

20240822_112716.jpg

## 데이터 시각화를 하는 이유
- 복잡한 데이터를 간단하게
    - 숫자로만 구성된 데이터는 이해하기 어렵지만, 시각적으로 표현하면 누구나 쉽게 파악
- 데이터 간의 관계 파악
    - 시각화를 통해 데이터 간의 관계를 직관적으로 확인
- 숨겨진 정보 발견
    - 숫자로만 분석하기 어려운 데이터 속에 숨겨진 패턴이나 이상치를 발견
- 효과적인 의사소통
    - 데이터를 시각화하면 복잡한 분석 결과를 간결하게 전달하고, 다른 사람들과 쉽게 공유

## 데이터 시각화의 활용 분야
- 비즈니스
    - 매출 분석, 고객 행동 분석, 마케팅 전략 수립 등
- 과학
    - 연구 결과 시각화, 데이터 분석
- 금융
    - 주식 시장 분석, 투자 포트폴리오 관리
- 의료
    - 환자 데이터 분석, 질병 진단
- 사회과학
    - 사회 현상 분석, 설문 조사 결과 시각화

## 데이터 시각화 도구
- Tableau
    - 데이터를 탐색하고 시각화하는 데 특화된 도구
- Python (Matplotlib, Seaborn)
    - 프로그래밍 방식으로 정교한 시각화를 구현
- 엑셀
    - 간단한 차트와 그래프를 생성하는 데 사용
- 파워 BI
    - 다양한 시각화 기능을 제공하며, 대규모 데이터 분석에 적합

# Matplotlib
- 파이썬 데이터 시각화의 기본 도구
- 파이썬에서 가장 널리 사용되는 데이터 시각화 라이브러리 중 하나
- 간단한 선 그래프부터 복잡한 맞춤형 시각화까지 다양한 형태의 그래프를 생성

## Matplotlib의 주요 특징
- 선 그래프, 막대 그래프, 산점도, 히스토그램, 파이 차트 등 다양한 종류의 플롯을 지원
- 플롯의 색상, 선 스타일, 마커, 레이블, 제목 등을 자유롭게 변경하여 원하는 시각화를 만들수 있음
- Jupyter Notebook 환경에서 바로 시각화 결과를 확인 가능

## Matplotlib의 장단점
- 장점
    - 유연성
    > 다양한 차트 종류와 커스터마이징 옵션을 제공하여 복잡한 시각화도 구현 가능
    - 확장성
    > 다른 라이브러리와의 연동이 용이하여 다양한 방식으로 활용
    - 오픈 소스
    > 무료로 사용할 수 있으며, 활발한 커뮤니티를 통해 많은 정보를 얻을 수 있음
- 단점
    - 복잡성
    > 기본적인 사용법은 간단하지만, 복잡한 시각화를 하려면 많은 코드를 작성해야 함
    - 학습 곡선
    > 다양한 기능을 익히는 데 시간이 걸릴 수 있음

## 언제 Matplotlib을 사용해야 할까?
- 간단한 시각화
    - 선 그래프, 막대 그래프 등 기본적인 시각화가 필요할 때
- 맞춤형 시각화
    - 다른 시각화 라이브러리로는 구현하기 어려운 복잡한 시각화가 필요할 때
- 다른 라이브러리와의 연동
    - 다른 시각화 라이브러리와 함께 사용하여 더욱 강력한 시각화를 구현하고 싶을 때

## figure와 axes 객체
- figure
    - 전체 그림판
    - Figure는 1개 이상의 그래프들을 담는 틀이나 창을 의미
    - 우리가 최종적으로 보게 되는 시각화 그래프들의 전체를 감싸는 가장 바깥쪽 영역
- axes
    - Axes는 Figure 내부에 존재하는 하나의 그래프 그리는 공간
    - 실제로 데이터를 시각화하여 그래프, 차트 등을 그리는 영역

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
plt.figure()
plt.axes()
plt.show() # 생략 가능, 명시 권장
````

## 단일 그래프

- plot 함수
    - line plot 을 그려주는 함수

````python
plt.plot(np.arange(2, 7))
plt.show()
````

## 다중 그래프
- 1개의 axes 안에 여러 개의 그래프 출력

````python
plt.figure(figsize=[15, 5])
plt.plot(np.arange(5))
plt.plot(np.arange(2, 7))
plt.show()
````

## subplots 함수를 이용해서 axes 여러 개 그리기

````python
fig, ax = plt.subplots(2,2,figsize=[15,5])
ax[0,0].plot(np.arange(5))
ax[0,1].plot(np.arange(2, 7))
ax[1,0].plot(np.arange(5))
ax[1,1].plot(np.arange(2, 7))
plt.show()
````

- 다른 방식

````python
plt.figure(figsize=[15,5])

plt.subplot(2, 2, 1)
plt.plot(np.arange(5))

plt.subplot(2, 2, 2)
plt.plot(np.arange(5))
plt.plot(np.arange(2, 7))

plt.subplot(223) # ,(콤마) 생략가능
plt.plot(np.arange(5))

plt.subplot(224)
plt.plot(np.arange(5))

plt.show()
````

## 타이틀 설정하기

````python
fig, ax = plt.subplots(1, 2, figsize=[15, 5])
fig.suptitle('제목')

x = np.arange(10)
ax[0].plot(x, np.exp(x))
ax[0].set_title("ax1 제목")

x = np.arange(1, 1000)
ax[1].plot(x, np.log(x))
ax[1].set_title("ax2 제목")

plt.show()
````

````python
x = np.arange(1, 1000)
y = np.log(x)
````

## x, y 축에 이름 넣기

````python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x축 이름')
ax.set_ylabel('y축 이름', rotation=0, labelpad=20)
plt.show()
````

## 눈금 회전하기

````python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x', rotation=45)
plt.show()
````

## 눈금 지정하기

````python
fig, ax = plt.subplots(figsize=[15,10])
ax.plot(np.arange(50))
ax.set_xticks(np.arange(50))
ax.set_yticks(np.arange(50))
plt.show()
````

## marker, line style, line color 변경하기

````python
fig, ax = plt.subplots(2, 2, figsize=[15,5])
ax[0,0].plot(np.exp(range(0, 10)), marker="o")
ax[0,1].plot(np.exp(range(0, 10)), marker="v", color="r", linestyle="--")
ax[1,0].plot(np.exp(range(0, 10)), marker="x", color="black", linestyle="")
ax[1,1].plot(np.exp(range(0, 10)), marker="^", color="#D1B2FF", linestyle=":")
plt.show()
````

## 범례 표시하기

````python
fig, ax = plt.subplots()

y = np.arange(10)
ax.plot(y)
ax.plot(y**2, alpha=0.2)
ax.legend(["Y", "Y^2"], loc=[0.03, 0.85])

plt.show()
````

# Box Plot
- 4분위수를 기준으로 데이터의 최대, 최소, Q1(25%지점), Q2(50%지점), Q3(75%지점) 등 통계적인 부분들을 손쉽게 시각화
- 이상치도 빠르게 판단이 가능
- IQR(Inter Quantile Range) : 3Q - Q1 , 3사분위수와 1사분위수의 차이
- Box Plot에서 최소값: Q1 - IQR * 1.5
- Box Plot에서 최대값: Q3 + IQR * 1.5
- Box Plot에서 최소값과 최대값을 벗어나면 이상치로 해석함
