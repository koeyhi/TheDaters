# 결측치 처리

```python
# 데이터프레임 내 결측치가 있는지 확인
df.isnull().sum()
```

```python
# 결측치 제거 -> 결측치가 있는 행 삭제
df_dropped_rows = df.dropna()

# 결과 출력
df_dropped_rows.head()
```

```python
# 결측치가 있는 열 삭제
df_dropped_columns = df.dropna(axis=1)

# 결과 출력
df_dropped_columns.head()
```

```python
# 'age' 컬럼의 결측치를 평균값으로 대체
df['eq_site_limit'].fillna(df['eq_site_limit'].mean(), inplace=True)

# 결과 확인
df.head()
```

```python
# 'age' 컬럼의 결측치를 중앙값으로 대체
df['eq_site_limit'].fillna(df['eq_site_limit'].median(), inplace=True)

# 결과 확인
df.head()
```

```python
# 'category_col'의 결측치를 최빈값으로 대체
df['eq_site_limit'].fillna(df['eq_site_limit'].mode()[0], inplace=True) # categorical 변수에 적합

# 결과 확인
df.head()
```

```python
from sklearn.impute import SimpleImputer

# 평균값으로 결측치 대체
imputer = SimpleImputer(strategy='mean')
df[['eq_site_limit']] = imputer.fit_transform(df[['eq_site_limit']]) # 선형 보간, machine learning 등의 방법론 가능

# 결과 확인
df.head()
```

# 이상치 제거

## 다양한 방법이 있으나, 가장 기본적인 것은 상/하위 threshold % 만큼의 데이터를 제거하는 것

```python
# 특정 컬럼의 상위 1% 값 출력하기 (예: 'eq_site_limit' 컬럼 사용)
percentile_99 = df['eq_site_limit'].quantile(0.99)  # 99번째 백분위수 값 계산

# 상위 1%에 해당하는 값만 필터링하여 출력
top_1_percent = df[df['eq_site_limit'] > percentile_99]

# 결과 출력
top_1_percent
```
