# Colab(코랩)
- Colab은 Google Colaboratory의 줄임말로, 구글에서 무료로 제공하는 클라우드 기반의 Jupyter 노트북 환경
- 복잡한 설치 과정 없이 브라우저만 있다면 언제 어디서든 강력한 컴퓨팅 자원을 이용하여 파이썬 코드를 실행하고 머신러닝 모델을 개발할 수 있는 환경을 제공

## Colab의 주요 특징
- 무료
    - 강력한 GPU를 포함한 Google 클라우드 컴퓨팅 자원을 무료로 사용
- 편의성
    - 브라우저 기반으로 작동하여 별도의 설치가 필요 없으며, Google Drive와 연동되어 파일 관리가 편리
- 공유 및 협업
    - 다른 사용자와 노트북을 공유하고 함께 작업
- 강력한 기능
    - 파이썬, TensorFlow, PyTorch 등 다양한 라이브러리를 지원하며, 대규모 데이터 처리 및 모델 학습이 가능
- GPU 지원
    - 고성능 GPU를 활용하여 딥러닝 모델을 빠르게 학습

# Colab 연결
- Colab에서는 ipynb 파일을 열고 클라우드 자원(컴퓨팅 자원)을 할당받기 위해 연결작업을 해야한다.
- 우측 상단에 연결 버튼을 클릭하면 된다.

# Colab 사용법

- `shift + enter`( 셀실행후 아래 셀선택)

- 자동완성은 `ctrl + space`를 입력하거나, 코드를 입력하는 과정에서 자동으로 된다.
- 도움말에 경우는 도움말을 보고자하는 함수등을 클릭하고 마우스를 가져다 놓으면 볼수 있다.

- 여러 셀 실행하기
    - 전체 셀 실행: `런타임` 메뉴에서 `모두 실행` 클릭
    - 현재 셀의 위에 있는 셀을 모두 실행: `런타임` 메뉴에서 `이전 셀 실행`클릭

- 노트북 파일 재시작 및 종료하기
    - 재시작: `런타임` 메뉴에서 `런타임 다시 시작` 클릭
    - 종료: `런타임` 메뉴에서 `런타임 연결 해제 및 삭제` 클릭
