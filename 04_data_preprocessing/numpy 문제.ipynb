## 1. MAE 와 RMSE 함수 구현하기
- 다음과 같이 실제 값과 , 모델이 예측한 값이 각각 100개가 있다.
- 실제값과 예측값 사이에 평균 오차를 구하는 함수 두개를 구현해보자.

````python
np.random.seed(42)
target = np.random.randint(50, 91,100) # 정답값
pred = np.random.randint(40, 100,100) # 예측값
target , pred
````

### MAE(Mean Absolute Error)
- 실제 값과 예측 값의 차이를 절대값으로 변환해 평균
$$
MAE = \frac{1}{n}{\sum_{i=1}^{n}|y_i-\hat{y}_i|}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def mae(target, pred):
    return np.mean(np.abs(np.subtract(target, pred)))

mae(target, pred)
````

### RMSE(Root Mean Squared Error)
- 실제 값과 예측 값의 차이를 제곱한 뒤 평균한 후에 루트

$$
RMSE = \sqrt{\frac{1}{n}{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def rmse(target, pred):
    return (np.mean(np.square(target - pred))) ** 0.5

rmse(target, pred)
````

````python
def split_dataset(dataset):
    data_size = len(dataset)
    row_size = int(len(dataset) * 0.2)
    selected_index = set(range(data_size))

    for _ in range(4):
        index_lst = np.random.choice(list(selected_index), row_size, replace=False)
        selected_index -= set(index_lst)
        yield index_lst

    yield list(selected_index)
````

````python
def split_dataset(dataset):
    data_size = len(dataset)
    row_size = int(len(dataset) * 0.2)
    indices = np.arange(data_size)
    np.random.shuffle(indices)

    for i in range(4):
        index_lst = indices[i * row_size : (i * row_size) + row_size]
        yield index_lst

    yield indices[4 * row_size :]
````

````python
def split_dataset(dataset):
    data_size = len(dataset)
    row_size = int(len(dataset) * 0.2)
    indices = np.arange(data_size)
    np.random.shuffle(indices)

    n_sub = data_size - row_size * 5
    index_list = indices[: data_size - n_sub].reshape(5, row_size).tolist()

    if n_sub > 0:
        index_list[-1].extend(indices[data_size - n_sub :].tolist())
    yield from index_list
````

### 딥러닝 모델에서 시계열 학습데이터 형태로 반환하는 함수를 만드시오.
- samsung_stock 변수의 데이터는 2차원 데이터이다.
- 입력데이터는 샘플별로 10일치의 행과 4열(시작가 ,상한가 , 하한가, 종가) 형태의 행렬을 담고 있는 3차원 형태의 데이터셋으로 변경하시오.
- 정답데이터는 미래의 5일치의 종가를 예측해야 한다는 가정하에 행별로 미래의 5일치의 종가가 담긴 2차원 형태의 데이터셋으로 변경하시오.
- 1일씩 이동하면서 데이터셋을 구성하시오.

```
아래 셀의 코드 작성하여 함수 구현
```

````python
samsung_stock
````

````python
def transform_data(samsung_stock, seq_len=10, pred_len=5):
    train, test = [], []
    for i in range(len(samsung_stock) - seq_len - pred_len + 1):
        train.append(samsung_stock[i : i + seq_len])
        test.append(samsung_stock[i + seq_len : i + seq_len + pred_len, 3])

    return np.array(train), np.array(test)
````

````python
def transform_data(samsung_stock, seq_len=10, pred_len=5):
    num_features = samsung_stock.shape[1]
    X = np.lib.stride_tricks.sliding_window_view(samsung_stock, (seq_len, num_features))[:-pred_len, 0, :, :]
    y = np.lib.stride_tricks.sliding_window_view(samsung_stock[:, 3], pred_len)[seq_len:]

    return X, y
````

- 다음 셀들은 위에 함수 실행 결과를 검증하는 부분

````python
x_train , y_train = transform_data(samsung_stock)
x_train.shape, y_train.shape # ((234, 10, 4), (234, 5))
````
