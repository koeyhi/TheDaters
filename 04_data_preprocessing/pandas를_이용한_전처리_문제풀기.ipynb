```
고유한 고객번호로만 이루어진 데이터프레임을 만들어서 features 변수에 넣으시오.
컬럼명은 ID로 하시오.
컬럼이 ID만 있는 데이터프레임을 만들라는 의미
중복 고객번호가 있으면 안된다.
```

````python
features = pd.DataFrame(df["ID"].unique().tolist(), columns=["ID"]).sort_values(by="ID")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 평균을 구하고 데이터 프레임 형식으로 정돈 하고 구매가격의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오.
```

````python
purchase_mean = df.groupby("ID")["구매가격"].mean().reset_index().rename(columns={"구매가격":"구매가격평균"})
purchase_mean
````

````python
features = pd.merge(features, purchase_mean, on="ID", how="left")
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최대값과 대분류의 고유값의 개수(다양성)를
데이데이터 프레임 형식으로 정돈 하고 구매가격과 대분류의 컬럼명을 변경하여
features 변수의 데이터프레임과 left merge 하시오
```

````python
agg_dict = {
    "구매가격": "max",
    "대분류": "nunique"
}
````

````python
tmp = df.groupby("ID").agg(agg_dict).rename(columns={"구매가격": "구매가격최대값", "대분류": "대분류고유값"})
features = pd.merge(features, tmp, on="ID", how="left")
````

````python
features
````

```
groupby 메소드를 이용하여 고객별로 구매가격의 최소값과, 총 구매액
그리고 지점코드의 고유값 개수와 최빈값을 구하시오
구하는 과정에서 집계 컬럼에 대한 컬럼명도 정의하시오.
데이터프레임형식으로 정돈하고 features 변수의 데이터프레임과 left merge 하시오.
```

````python
df.columns
````

````python
agg_dict = {
    "구매가격": [
        ("구매가격최소값", "min"),
        ("구매가격합계", "sum")
    ],
    "지점코드": [
        ("지점코드고유값", "nunique"),
        ("지점코드최빈값", lambda x: x.mode())
    ],
}
````

````python
tmp = df.groupby("ID").agg(agg_dict)
tmp.columns = tmp.columns.droplevel(0)
features = pd.merge(features, tmp, on="ID", how="left")
features
````
