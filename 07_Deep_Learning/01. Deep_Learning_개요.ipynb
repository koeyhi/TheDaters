# 딥러닝 개요

## 인공지능의 개념 이해
- 인공지능(AI)은 기계가 인간과 유사한 인지적 기능을 수행할 수 있도록 만드는 기술
- AI는 다양한 하위 분야로 나뉘며, 대표적인 예로는 머신러닝(ML)과 딥러닝(DL)이 있음

## 머신러닝과 딥러닝의 차이
- 머신러닝
> 데이터를 이용해 패턴을 학습하고 예측하거나 의사 결정을 내리게 하는 기술
- 딥러닝
> 머신러닝의 하위 분야로, 인공 신경망(Artificial Neural Networks)을 기반으로 더 깊고 복잡한 모델을 만들어 데이터를 학습


## 신경망 (Neural Network)
- 생물학적 신경망(Biological Neural Network, 신경회로)
    - 인간과 동물의 뇌에서 발견되는 복잡한 신경 세포들의 네트워크
    - 일정 크기의 자극(신호)를 입력받아 활성화됨으로 특정 기능을 수행하는 시냅스로 연결된 신경 집단(Neural Group)
    - 뉴런이라는 기본 단위로 구성
    - 감각 정보의 수집, 처리, 전달을 통해 생각, 기억, 학습, 의사결정 등을 수행
- 인공 신경망 (Artificial Neural Network)
    - 생물학적 신경망을 모방하여 컴퓨터 시스템 상에서 구현된 알고리즘 구조
    - 생물학적 신경망에서 영감을 얻어 만들어진 수학적 모델링 알고리즘으로 생물학적 신경망의 작동방식과 구조가 유사


### 뉴런의 개념
- 생물학적 신경망
    - 뉴런은 세포체, 가지돌기(수상돌기), 축삭으로 구성
    - 뉴런은 시냅스를 통해 다른 뉴런과 연결되고, 전기적 신호와 화학적 신호를 통해 정보를 전달
- 인공 신경망
    - 인공 뉴런은 수학적 함수의 형태로, 입력값을 받아 가중치와 바이어스를 적용한 후 활성화 함수를 통해 출력

### 인공 뉴런
- 뉴런을 모방하는 기초단위 노드(Node)
    - 각각의 신경 단위에서 많은 입력들을 조합해서 하나의 출력값으로 배출
    - 비선형 변환(활성함수)을 통해 다음 노드에 전달
- w: weight(가중치)
- b: bias(편향)
- h: Activation Function

$$
a = b + w_1x_1 + w_2x_2 + ... + w_mx_m
$$
$$
y = h(a)
$$

### 뉴런 연결 방식
- 생물학적 신경망
    - 수백억 개의 뉴런이 복잡한 형태로 서로 연결
    - 각 뉴런은 여러 다른 뉴런들과 시냅스를 통해 다중 연결되어 있음
- 인공 신경망
    - 인공 신경망은 계층적 구조(입력층, 은닉층, 출력층)를 갖고있음
    - 각 뉴런은 층 내에서 다른 뉴런과 연결되며, 연결 강도를 결정하는 가중치가 부여
    - 다음의 그림에서 hidden layer를 완전연결계층(Fully-Connected Layer) 라고 함

## 신경망 학습 과정
1. 전방 전파(Forward Propagation)
    - 입력 데이터가 네트워크를 통해 출력으로 전달되며 각 뉴런의 가중치와 바이어스에 의해 연산이 이루어진다.
2. 손실 함수
    - 출력과 실제 값의 차이를 계산해 손실(Loss)을 구한다.
    - 손실 함수는 회귀 문제에서는 평균 제곱 오차(MSE), 분류 문제에서는 교차 엔트로피(Cross Entropy) 등이 사용된다.
3. 역전파(Backpropagation)
    - 손실 함수를 최소화하기 위해 가중치와 바이어스를 조정하기 위해 가중치의 경사를 업데이트하는 과정이다.
4. 최적화(Optimization)
    - 경사 하강법(Gradient Descent)을 통해 최적의 가중치 값을 찾아간다.

## 딥러닝(Deep Learning)
- 딥러닝은 머신러닝 알고리즘 중 하나인 인공신경망을 다양하게 쌓은 것
- 딥러닝은 머신러닝이 처리하기 어려운 데이터(비정형 데이터)를 더 잘 처리
- 다음의 그림과 같은 형태의 신경망을 피드 포워드(Feed-forward) 신경망이라고 함
    - 입력에서 출력으로 이어지는 과정이 한방향으로 흘러 순환이 없는 신경망

20220918123710.png

### 딥러닝은 만능?
- 학습을 위해 상당히 많은 양의 데이터를 필요로 함
    > 딥러닝은 많은 양의 데이터를 필요로 한다. 데이터가 부족할 경우 과적합(overfitting) 문제가 발생할 수 있음
- 계산이 복잡하고 수행시간이 오래 걸림
    > 딥러닝 모델의 학습에는 높은 연산 능력이 필요하다. 이를 위해 GPU 같은 하드웨어 가속기를 사용
- 이론적 기반이 없어 결과에 대한 장담이 어려움
- 블랙 박스 접근 방식

## 경사하강법(Gradient Descent)
- 모델이 잘 학습할수 있도록 기울기(변화율)을 사용하여 모델의 파라미터를 조정하는 방법
    1. 예측과 실제값을 비교하여 손실을 구한다.
    2. 손실이 작아지는 방향으로 파라미터를 수정한다.
    3. 이과정을 반복한다.
    - 경사(기울기): 파라미터에 대한 오차의 변화
        - 기울기 + : 파라미터 ↓
        - 기울기 - : 파라미터 ↑

## 순전파(forward propagation)
- Forward pass
- 입력 데이터가 신경망을 통해 전달되면서 예측 결과가 출력되는 과정

## 역전파(Back propagation)
- backward pass
- 신경망의 학습 과정에서 중요한 알고리즘으로, 가중치와 바이어스를 조정하여 모델의 성능을 점진적으로 향상시키는 역할
- 효율적인 계산을 위해 역전파 알고리즘을 사용
    - 손실값을 구해 이 손실에 관여하는 가중치들의 경사를 손실이 작아지는 방향으로 수정하는 알고리즘
    - 순전파 후 예측값과 실제값의 차이의 기울기 계산에 따른 가중치의 경사를 업데이트하는 과정
- 신경망의 모든 가중치의 국소적인 미분을 전달하는 원리는 연쇄법칙(chain rule)을 따름
- 합성함수
    - 여러 함수로 구성된 함수
- chain rule
    - 합성함수의 미분의 성질
    - 합성함수의 미분은 합성함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있음

$$
y = f(z) , z = g(x)
$$

$$
y = f(g(x))
$$



$$
\frac{dy}{dx} = \frac{dy}{dz} * \frac{dz}{dx}
$$


- 역전파 과정의 단계
    1. 출력층에서 오차를 계산하고, 이 오차를 이용해 출력층의 가중치에 대한 기울기를 구한다.
    2. 출력층의 오차가 구해지면, 이 정보를 사용해 바로 이전 층(은닉층)의 기울기를 계산한다.
    > 이를 반복하여 입력층에 이를 때까지 기울기를 구한다.
    3. 모든 층에서 기울기가 계산되면, 가중치를 업데이트한다.
    > 일반적으로 경사 하강법 알고리즘을 사용하여 가중치와 바이어스를 원하는 방향으로 조정한다.

## 손실함수(Loss Function)
- 모델의 출력값(Output)과 정답과의 차이(오차,Error)를 의미
- 신경망이 학습할 수 있도록 해주는 지표
- 손실 값이 최소화 되도록 하는 가중치(weight)와 편향(bias)를 찾는 것이 학습의 목표

## 배치(Batch)
- 일을 모아서 한번에 처리하는 단위
- 배치의 사전적 의미는 (일괄적으로 처리되는) 집단이다. 배치는 한번에 여러개의 데이터를 묶어서 입력하는 것


- Batch Gradient Descent
    - 모든 데이터를 사용하여 손실함수의 Gradient 계산
    - 장점: 전체 데이터에 대해 error gradient를 계산하기 때문에 optimal로 수렴이 안정적이다.
    - 단점: local optimal 상태가 되면 빠져나오기 힘듦


- Stochastic Gradient Descent
    - 확률적 표본 데이터를 사용하여 손실함수의 Gradient 계산
    - 장점: 아래 이미지에서 보이듯이 shooting이 일어나기 때문에 local optimal에 빠질 위험이 적다.
    - 단점: global optimal을 찾지 못 할 가능성이 있다.


- Mini-Batch Gradient Descent*
    - 일정 크기의 표본 데이터를 사용하여 손실함수의 Gradient 계산
    - 장점: BGD보다 local optimal에 빠질 리스크가 적다.
    - 단점: batch size(mini-batch size)를 설정해야 한다.

## Optimizer
> 미니배치 경사하강법의 방식을 보완하고, 학습 속도를 높이기위한 알고리즘

- SGD(확률적 경사 하강법)
  - 랜덤하게 추출한 일부 데이터를 사용해 더 빨리 업데이트 하게 하는 것
  - 효과: 속도 개선
- Momentum
    - 기존 업데이트에 사용했던 경사의 일정 비율을 남겨서 현재의 경사와 더하여 업데이트함
    - 효과: 정확도 개선
- Adagrad
    - 각 파라미터의 update 정도의 따라 학습률의 크기를 다르게 해줌
    - 효과: 보폭 크기 개선
- RMSProp
    - Adagrad의 경우 update가 지속됨에 따라 학습률이 점점 0에 가까워지는 문제가 발생
    - 이전 update 맥락을 보면서 학습률 조정하여 최신 기울기를 더 크게 반영
    - 효과: 보폭 크기 개선
- Adam
    - Momentum 과 RMSProp 장점을 함께 사용
    - 가장 많이 사용하는 Optimizer
    - 효과: 정확도와 보폭 크기 개선
