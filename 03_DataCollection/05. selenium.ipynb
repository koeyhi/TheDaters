# 렌더링(Rendering) 이란?
- 객체(Object)를 2차원 화면인 하나의 장면으로 바꾸어 표현하는 것을 말함
- 서버로부터 넘겨받은 HTML 코드는 웹 브라우저에서 읽어온 후 화면에 시각적으로 표시하게 되는데, 이 과정을 렌더링이라고 함
- 화면에 표시할 웹 페이지를 만드는 과정

# SSR(Server Side Rendering)과 CSR(Client Side Rendering)
- SSR
    - 서버로부터 완전하게 만들어진 HTML을 받아와 페이지 전체를 렌더링 하는 방식
    - 전통적인 웹 어플리케이션 렌더링 방식으로 사용자가 웹 페이지에 접근할 때, 서버에 페이지에 대한 요청
    - 참고로 웹 서버에 요청할 때마다 브라우저에서 새로고침이 일어나고 서버에 새로운 페이지에 대한 요청을 하는 방식
- CSR
    - 사용자 요청에 따라 필요한 데이터만 응답받아 클라이언트측 브라우저에서 렌더링 하는 방식
    - HTML 및 static 파일들이 로드된다. 이때, 데이터가 있다면 데이터 또한 서버에 요청하여 화면에 나타낸다.
    - 브라우저가 서버에 HTML과 static 파일을 요청한 후 로드되면, 사용자의 상호작용에 따라 JavaScript를 통해 동적으로 렌더링을 한다.

## Selenium의 주요 기능
- 브라우저 제어
    - 크롬, 파이어폭스, 엣지 등 다양한 브라우저를 프로그램에서 직접 제어
- 웹 페이지 조작
    - 웹 페이지의 요소를 찾아 클릭, 입력, 스크롤 등 다양한 조작을 수행
- JavaScript 실행
    - 웹 페이지에 포함된 JavaScript 코드를 실행
- 웹 애플리케이션 테스트
    - 웹 애플리케이션의 기능을 자동으로 테스트하여 오류를 찾아낼 수 있음

## Selenium을 사용하는 이유
- 복잡한 웹 사이트 자동화
    - JavaScript를 많이 사용하거나 동적으로 변하는 웹 페이지도 자동화할 수 있음
- 웹 스크래핑
    - 정적인 HTML뿐만 아니라 동적으로 로딩되는 콘텐츠도 효과적으로 추출
- 웹 테스트
    - 웹 애플리케이션의 기능을 자동으로 테스트하여 개발 효율을 높일 수 있음

## Selenium 3.0과 4.0 달라진 점
- 3.0 버전에서는 수동으로 브라우저의 드라이버를 다운로드하고 경로를 지정해줘야 했음.
    - 단점으로 os 버전에 맞는 드라이버를 다운로드해야함.
    - 브라우저 업데이트시 버전에 맞는 드라이버를 다시 다운로드해야하는 번거로움
- 4.0 부터는 WebDriver Manager 가 자동으로 Web browser의 버전을 확인하고 OS와 버전에 맞는 webdriver를 다운로드하고 실행

# 웹페이지가 로딩이 될때까지 대기
- time 모듈의 sleep 함수는 지정한 시간까지 무조건 대기하지만, 셀레니움의 implicitly_wait 메서드를 사용하면 웹페이지가 로딩이 완료되면 대기를 종료하고 다음 동작으로 넘어가게 한다.

````python
url = 'https://www.naver.com'
````

````python
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(3000)  # 초단위
driver.quit()
````

# 엘리먼트 선택 후 클릭 또는 입력하기
- driver 객체의 find_element 메서드
    - 크롬 브라우저에 로딩된 요소를 선택하여 해당 엘리먼트 객체를 반환
    - 첫번째 인수로 어떤방식으로 요소를 선택할지 지정
    - 두번째 인수로 해당 방식에 맞는 선택자 넣으면 됨

- driver 객체의 find_elements 메서드
    - 크롬 브라우저에 로딩된 요소들을 선택하여 해당 엘리먼트 객체들을 리스트에 담아 반환
    - 첫번째 인수로 어떤방식으로 요소를 선택할지 지정
    - 두번째 인수로 해당 방식에 맞는 선택자 넣으면 됨

- 엘리먼트 객체의 click 메서드
    - 해당 엘리먼트를 클릭

- 엘리먼트 객체의 send_keys 메서드
    - 해당 엘리먼트에 값 입력

````python
By.CSS_SELECTOR # css 선택방식
````

````python
By.XPATH # xpath 선택방식
````

````python
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.implicitly_wait(5)

driver.find_element(By.CSS_SELECTOR, "#query").send_keys("텐서플로우")
driver.find_element(By.CSS_SELECTOR, "#search-btn").click()

time.sleep(5)

driver.quit()
````

````python
from selenium.webdriver.common.keys import Keys
````

````python
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.implicitly_wait(5)

el = driver.find_element(By.CSS_SELECTOR, "#query")
# el.send_keys('텐서플로우')
# el.send_keys(Keys.ENTER)
el.send_keys("텐서플로우", Keys.ENTER)

time.sleep(5)

driver.quit()
````

````python
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.implicitly_wait(5)

el_list = driver.find_elements(By.CSS_SELECTOR, ".shortcut_list .shortcut_item")
for el in el_list:
    time.sleep(2)
    el.click()

time.sleep(5)

driver.quit()
````

# 새 탭 열기 및 탭 이동

````python
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.implicitly_wait(5)

el = driver.find_element(By.CSS_SELECTOR, "#special-input-logo > a").send_keys(
    Keys.CONTROL, Keys.ENTER
)

driver.switch_to.window(driver.window_handles[1])  # 두 번째 탭으로 이동
time.sleep(3)

driver.switch_to.window(driver.window_handles[0])  # 첫 번째 탭으로 이동
driver.close()  # 탭 닫기

time.sleep(5)

driver.quit()
````

````python
driver = webdriver.Chrome(options=options)
driver.get(url)
driver.implicitly_wait(5)

scripts = f'window.open("{url}");'
driver.execute_script(scripts)

driver.switch_to.window(driver.window_handles[1])  # 두 번째 탭으로 이동
time.sleep(3)

driver.switch_to.window(driver.window_handles[0])  # 첫 번째 탭으로 이동
driver.close()  # 탭 닫기

time.sleep(5)

driver.quit()
````
