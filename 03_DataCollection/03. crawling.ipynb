# 웹 크롤링(Web Crawling)
- 복수 개의 웹페이지를 탐색하는 행위
- 자동으로 웹페이지의 정보를 수집하는 일
- 방식
    - 시작 URL을 설정하고, HTML 코드를 분석하여 링크를 추출
    - 추출된 링크를 따라 새로운 페이지를 요청하고, 이 과정을 반복
    - 수집된 데이터를 저장하거나 처리

## 웹 스크래핑(Web Scraping)
- 웹 페이지에서 특정 부분의 필요한 데이터만을 추출하는 과정
- 크롤링을 통해 수집한 정보를 분석해서 필요한 정보만 추출하는 일
- 방식
    - 수집된 HTML 문자열을 HTML 코드 형태로 변환
    - 원하는 데이터를 추출하기 위한 HTML 요소에 접근할 수 있는 방법을 사용(예: CSS 선택자, XPath)
    - 추출된 데이터를 정제하고 저장

## 크롤링과 스크래핑의 차이점

||크롤링|스크래핑|
|------|---|---|
|목적|웹 페이지 탐색 및 수집|특정 데이터 추출|
|범위|전체 웹|특정 웹 페이지|
|데이터|링크, HTML|구체적인 내용(텍스트, 숫자 등)|


## 웹 크롤링 주요 용도
- 웹 크롤링은 다양한 분야에서 활용
- 시장 연구, 경쟁사 분석, SNS 트랜드 파악 등 다양한 분야에서 크롤링을 활용해 데이터를 수집하고 분석하여 의사결정에 활용

## 활용 사례
- E-commerce 사이트에서의 가격 정보 수집
    - 온라인 쇼핑몰에서는 상품의 가격 정보를 주기적으로 업데이트한다.
    - 크롤링을 통해 이러한 가격 변동을 실시간으로 파악하고, 경쟁사의 가격 전략을 분석하는데 활용될 수 있다.
    - 이는 마케팅 전략 수립에 있어 중요한 역할을 한다.
- 소셜 미디어 데이터 분석을 위한 크롤링
    - 소셜 미디어는 현재의 트렌드와 사회의 의견을 반영하는 중요한 공간이다.
    - 특정 주제나 키워드에 대한 사람들의 반응을 파악하기 위해 크롤링을 통한 데이터 수집이 이루어진다.
    - 이렇게 수집된 데이터는 다양한 분석 방법을 통해 유용한 인사이트를 제공한다.

# HTML(Hyper Text Markup Language)
- 웹 페이지에 필요한 구성요소를 구성하는 Markup 언어
- 태그를 이용해 화면의 레이아웃 구조와 텍스트, 요소 등을 표현


```html
<!DOCYTYPE HTML>
<html>
    <head>
        <title>웹문서의 제목</title>
        <meta charset="utf-8">
    </head>
    
    <body>
        웹문서의 내용(실제 브라우저에 표시되는 내용)
    </body>
</html>
```



### HTML Element
- HTML 요소는 HTML 문서나 웹 페이지를 이루는 개별적인 요소를 의미
- 태그에 이름을 지정하는 속성
    - `id`
        - 태그에 유일한 이름을 붙이고 싶을 때 사용
        - `id` 속성의 값은 전체 페이지에서 단 하나의 요소에만 지정이 가능 (문법상 강제는 아님, 오류 x)
    - `class`
        - 여러 태그에 동일한 이름을 붙이고 싶을 때 사용
        - `class` 속성 값은 전체 페이지에서 여러 요소에 동일한 값으로 지정이 가능
        - `id` 속성과는 달리 하나의 요소에 여러개의 `class` 이름을 지정할 수 있음
            > ex: `<div class="cls1 cls2"></div>`

# CSS(Casading Style Sheet)
- HTML 의 요소를 꾸며주는 스타일 시트
- HTML 문서에 있는 요소들에 대하여 스타일을 적용
- css 선택자를 이용하여 html 요소에 접근하여 스타일을 적용
- css 선택자
    - 태그명으로 접근하는 예시
        - a 태그에 접근 : `a`
    - 태그의 class 속성명으로 접근하는 예시
        - class="btn" : `.btn`
    - 태그의 id 속성명으로 접근하는 예시
        - id="btn" : `#btn`

- p 태그의 텍스트를 파란색으로 적용하는 예시
```CSS
p {
    color: blue;
}
```

````python
%%html
<!DOCTYPE html>

<html>
    <head>
        <meta charset="UTF-8">
        <title>HTML & CSS 연습</title>

        <style type="text/css">
        /* 나는 css 주석 입니다. */

        h2{
            border-bottom: 3px solid #000;
        }
        .box{padding: 20px;border-bottom: 1px solid red;}

        #css-doc-link{display: inline-block;border: 2px dotted blue;}

        </style>

    </head>

    <body>
        <!-- 나는 HTML 주석 입니다. -->

        <!-- h 태그는 제목을 표시할 때 사용 하는 태그 h1 ~ h6 까지 있음 -->
        <h1>나의 HTML 페이지</h1>
        <h2>태그를 소개합니다.</h2>


        <!-- div 태그는 레이아웃을 나누는데 주로 쓰이는 태그 -->
        <div class="box">
            <h3 style="color:blue">하이퍼링크만들기</h3>

            <!-- a 태그는 링크를 걸어주는 태그 -->
            <a href="https://naver.com" target="_blank">네이버로 이동하기</a><br>
            <a href="https://www.google.com/" target="_blank">구글로 이동하기</a><br>
        </div>

        <div class="box">
            <h3 class="mytitle">목록 만들기</h3>

            <!-- ul 태그는 목록을 만드는 태그 -->
            <ul>
                <li class="mylang">넘파이</li>
                <li class="mylang">판다스</li>
                <li class="mylang">사이킷런</li>
                <li class="mylang">파이토치</li>
            </ul>
        </div>

        <div class="box">
            <h3>이미지 랜더링</h3>

            <!-- img 태그는 이미지를 삽입하는 태그 -->
            <img src="https://blog.velsoft.com/wp-content/uploads/2018/10/HTML-Basic.jpg">
            <br> <!-- 줄바꿈 태그 -->
            <a id="css-doc-link" href="https://developer.mozilla.org/ko/docs/Web/CSS" target="_blank">
                <img src="https://miro.medium.com/v2/resize:fit:600/1*OFsc0SD55jhi8cjo7aCA4w.jpeg">
            </a>
        </div>

        <div class="box">
            <h3>문단 만들기</h3>

            <!-- p 태그는 문단을 만드는 태그 -->
            <p>
                기계 학습 또는 머신 러닝은 경험을 통해 자동으로 개선하는 컴퓨터 알고리즘의 연구이다.<br>
                인공지능의 한 분야로 간주된다. <br>
                컴퓨터가 학습할 수 있도록 하는 알고리즘과 기술을 개발하는 분야이다. <br>
                가령, 기계 학습을 통해서 수신한 이메일이 스팸인지 아닌지를 구분할 수 있도록 훈련할 수 있다.<br>
            </p>
            <p>
                심층 학습 또는 딥 러닝은 여러 비선형 변환기법의 조합을 통해 높은 수준의 추상화를 시도하는 기계 학습 알고리즘의 집합으로 정의되며,<br>
                큰 틀에서 사람의 사고방식을 컴퓨터에게 가르치는 기계학습의 한 분야라고 이야기할 수 있다.
            </p>
        </div>

    </body>
</html>
````

# xpath(XML Path Language)
- XML 이나 HTML 의 특정 요소를 선택할때 사용
- xpath 사용법
    - `//` 로 시작해야 함
        - 하위 모든 엘리먼트를 찾겠다는 의미
        - `'div a'` => `'//div//a'`
    -  `/` : 바로 하위 엘리먼트를 찾겠다는 의미.
        - `'div > a'` => `'//div/a'`
    -  `*` : 모든 태그라는 의미
        - `'div > *'` => `'//div/*'`
    -  `[@id="idname"]`
        - `id`가 `idname`인 엘리먼트
    - `div[3]`
        - 같은 레벨의 div 태그중 에서 3 번째 엘리먼트를 선택
        - 1부터 시작

- lxml 라이브러리를 이용하면 xpath 로 접근 가능
- 설치

```bash
pip install lxml
```

````python
from lxml.html import fromstring
````

````python
xpath_obj = fromstring(html_str)
xpath_obj
````

````python
xpath_obj.xpath('//html/title') # html > title
````

````python
el = xpath_obj.xpath('//html//title') # html title
el
````

````python
el[0].text
````

````python
xpath_obj.xpath("//*[@class='box']") # .box
````

````python
xpath_obj.xpath("//div[@class='box']") # div.box
````

````python
el = xpath_obj.xpath('//a[@id="css-doc-link"]') # a#css-doc-link
el
````

- 요소의 속성값들을 가져오기

````python
el[0].attrib
````

````python
el[0].get("href")
````

````python
xpath_obj.xpath('//ul/li[3]')[0].text
````
