# 통신 프로토콜
- 통신 프로토콜 또는 통신 규약은 컴퓨터나 원거리 통신 장비 사이에서 메시지를 주고 받는 양식과 규칙의 체계
- 주요 프로토콜 예시
    - HTTP : Hyper Text Transfer Protocol
    - HTTPS : Hyper Text Transfer Protocol Secure
    - FTP : File Transfer Protocol
    - SMTP : Simple Mail Transfer Protocol
    - SSH : Secure Shell

# HTTP(Hyper Text Transfer Protocol) 란?
- HTML, Plain text, JSON, XML , 이미지 , 동영상 등 다양한 형태의 정보 전송하는 프로토콜
- HTTP는 서버/클라이언트 모델
- 클라이언트에서 요청(request)를 보내면 서버는 요청을 처리해서 응답(response)
- 클라이언트
    - 서버에 요청하는 클라이언트 소프트웨어(IE, Chrome, Firefox, Safari ...)가 설치된 컴퓨터를 말함
    - 클라이언트는 URL을 이용해서 서버에 접속하고, 데이터를 요청
- 서버
    - 클라이언트의 요청을 받아서, 요청을 해석하고 응답을 하는 소프트웨어가 설치된 서버측 컴퓨터
    - 웹서버는 보통 표준 포트인 80번 포트로 서비스 함

## HTTP 프로토콜 특징
- Connectionless 프로토콜(비연결 지향)
- 클라이언트가 서버에 요청(Request)을 했을 때, 그 요청에 맞는 응답(Response)을 보낸 후 연결을 끊는 처리방식

## HTTP Packet
- HTTP 통신으로 서버와 클라이언트 사이에 주고 받는 데이터를 말함
- Request Header
    - 클라이언트가 서버에 요청할 때 존재하는 헤더로 요청한 URL, Method, 브라우저 정보 등이 포함
- Respone Header
    - 서버가 클라이언트의 요청에 대한 응답을 할 때 존재하는 헤더로 콘텐츠에 사용된 인코딩, 응답을 생성하기 위해 서버 시스템에서 사용되는 서버 소프트웨어 및 기타 정보 등이 포함
- HTTP Body
    - 서버에서 응답시 클라이언트가 요청한 실제 데이터가 포함되어 있다.
    - HTML , image, css , javascript, text ...

## HTTP 메서드(Method)
- 서버에 요청을 보내는 방식
- GET 방식
    - URL 에 데이터를 담아 요청
- POST 방식
    - GET 방식과는 달리 body영역에 데이터를 담아 요청

## HTTP 상태코드(응답)
- 클라이언트가 서버에 요청을 하면, 서버는 요청에 대한 처리 상태를 숫자로 반환하는데 이를 상태 코드라고 한다.
- 상태 코드는 3자리 숫자로 만들어져 있으며, 첫번째 자리는 1에서 5까지 제공
    - 1xx: 요청을 받았으며 프로세스를 계속 진행
    - 2xx: 요청성공(ex. 200)
    - 3xx: 요청이 유효하지만 주소가 변경(ex. 301)
    - 4xx: 클라이언트 오류(ex. 404)
    - 5xx: 서버 오류(ex. 500)
    - https://developer.mozilla.org/ko/docs/Web/HTTP/Status

# Open API 란?
- 클라이언트가 특정 소프트웨어나 애플리케이션의 기능에 접근할 수 있도록 공개된 인터페이스를 의미
- 다양한 데이터를 어디서나 쉽게 이용할 수 있도록 개방된 API
- API(application programming interface)
    - 프로그램의 특정한 부분에 요청해서 그 안에 있는 데이터와 서비스(기능)를 이용할 수 있게 해주는 소프트웨어 인터페이스
    - 소프트웨어 간의 상호작용을 가능하게 하는 매개체

## Open API 주요 특징
- 공개 접근성
    - 누구나 접근할 수 있도록 공개되어 있어, 특정 소프트웨어의 기능을 다른 애플리케이션에 통합하거나 확장
- 표준화된 방식
    - 대부분의 Open API는 표준화된 방식을 사용하여, 데이터 전송과 처리에 있어 일관성을 유지
- 개발자 친화적
    - Open API는 보통 잘 정리된 문서와 샘플 코드가 제공되어, 개발자가 쉽게 이해하고 사용할 수 있도록 도와
- 확장성
    - Open API를 통해 외부 개발자들은 기존 애플리케이션의 기능을 확장하거나 새로운 기능을 추가
- 협업 촉진
    - 다양한 개발자들이 Open API를 사용하여 상호작용할 수 있으므로, 혁신적인 서비스나 애플리케이션이 등장할 가능성 증가

## 오픈 API 활용 분야
- 웹 서비스 개발
    - 다양한 웹 서비스를 개발하는 데 필요한 기능과 데이터를 활용
- 앱 개발
    - 모바일 앱 개발 시 지도, 검색, 결제 등 다양한 기능을 구현하는 데 활용
- 데이터 분석
    - 수집된 데이터를 분석하여 새로운 가치를 창출
- 머신 러닝
    - 머신 러닝 모델 학습에 필요한 데이터를 확보
