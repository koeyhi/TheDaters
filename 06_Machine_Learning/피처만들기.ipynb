# 특성 공학(Feature Engineering)

## 날짜 형식으로 변환

```python
train_tr["구매일시"] = pd.to_datetime(train_tr["구매일시"])
test_tr["구매일시"] = pd.to_datetime(test_tr["구매일시"])
```

## 새로 만든 feature와 병합할 고객ID로만 이루어진 데이터프레임 생성

```python
train_ft = train_target[["ID"]]
test_ft = submit[["ID"]]

train_ft.shape, test_ft.shape
```

## 구매일시를 이용한 특성생성

```python
agg_list = [
    # 컬럼명, 집계 방식
    ("내점일수", lambda x: x.dt.date.nunique()),
    ("구매주기", lambda x: int((x.max() - x.min()).days / x.dt.date.nunique())),
    # ("주중방문비율", lambda x: np.mean(x.dt.weekday <= 4)),
    ("주말방문비율", lambda x: np.mean(x.dt.weekday > 4)),
    ("주구매요일", lambda x: x.dt.weekday.mode()[0]),
    ("주구매시간대", lambda x: x.dt.hour.mode()[0]),
    ("일별평균구매건수", lambda x: x.count() / x.dt.date.nunique()),
    ("거래개월수", lambda x: x.dt.date.astype(str).str[:-3].nunique()),
]

tmp = train_tr.groupby("ID")["구매일시"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매일시"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 지점을 이용한 특성생성

```python
agg_list = [
    ("방문지점수", "nunique"),
    ("주구매지점", lambda x: x.mode()[0]),
]

tmp = train_tr.groupby("ID")["지점코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["지점코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 브랜드코드를 이용한 특성생성

```python
agg_list = [
    ("브랜드코드_nunique", "nunique"),
    # ("최다구매브랜드", lambda x: x.mode()[0])
]

tmp = train_tr.groupby("ID")["브랜드코드"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how='left',on="ID")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["브랜드코드"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left',on="ID")
test_ft.head()
```

## 중분류를 이용한 특성생성

```python
agg_list = [
    ("중분류_nunique", "nunique"),
    ("주구매_중분류", lambda x: x.mode()[0]),
]
tmp = train_tr.groupby("ID")["중분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["중분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 대분류를 이용한 특성생성

```python
agg_list = [
    ("대분류_nunique", "nunique"),
    ("주구매_대분류", lambda x: x.mode()[0]),
]

tmp = train_tr.groupby("ID")["대분류"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["대분류"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 가정용품 구매비율

```python
train_tr["가정용품구매비율"] = train_tr["대분류"].str.contains("가정")
tmp = train_tr.groupby("ID")["가정용품구매비율"].mean().reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
test_tr["가정용품구매비율"] = test_tr["대분류"].str.contains("가정")
tmp = test_tr.groupby("ID")["가정용품구매비율"].mean().reset_index()
test_ft = test_ft.merge(tmp, how="left")
test_ft.head()
```

## 아동제품 구매비율

```python
train_tr["아동용품구매비율"] = train_tr["대분류"].str.contains("아동|영")
tmp = train_tr.groupby("ID")["아동용품구매비율"].mean().reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
test_tr["아동용품구매비율"] = test_tr["대분류"].str.contains("아동|영")
tmp = test_tr.groupby("ID")["아동용품구매비율"].mean().reset_index()
test_ft = test_ft.merge(tmp, how="left")
test_ft.head()
```

## 공산품, 생식품 구매비율

```python
train_tr["공산품_생식품_구매비율"] = train_tr["대분류"].str.contains("공산품|생식")
tmp = train_tr.groupby("ID")["공산품_생식품_구매비율"].mean().reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
test_tr["공산품_생식품_구매비율"] = test_tr["대분류"].str.contains("공산품|생식")
tmp = test_tr.groupby("ID")["공산품_생식품_구매비율"].mean().reset_index()
test_ft = test_ft.merge(tmp, how="left")
test_ft.head()
```

## 구매가격을 이용한 특성생성

```python
agg_list = [
    ("총구매액", lambda x: x[x > 0].sum()),
    ("구매건수", "count"),
    ("평균구매액", lambda x: x[x > 0].mean()),
    ("최대구매액", lambda x: x[x > 0].max()),
    # ("최소구매액", lambda x: x[x > 0].min()),
    ("환불금액", lambda x: x[x < 0].sum()),
    ("환불건수", lambda x: (x < 0).sum()),
    ("구매금액표준편차", lambda x: x[x > 0].std()),
    ("구매금액왜도", lambda x: x.skew()),
    ("구매금액첨도", lambda x: x.kurtosis()),
]

tmp = train_tr.groupby("ID")["구매가격"].agg(agg_list).reset_index()
train_ft = train_ft.merge(tmp, how="left")
train_ft.head()
```

```python
tmp = test_tr.groupby('ID')["구매가격"].agg(agg_list).reset_index()
test_ft = test_ft.merge(tmp, how='left')
test_ft.head()
```

## 최다구매브랜드별 구매금액

```python
tmp = train_tr.groupby(["ID", "브랜드코드"])["구매가격"].sum().reset_index()
tmp = tmp.loc[tmp.groupby("ID")["구매가격"].idxmax()]
tmp.columns = ["ID", "최다브랜드코드", "최다브랜드금액"]
train_ft = train_ft.merge(tmp, how="left", on="ID")
train_ft
```

```python
tmp = test_tr.groupby(["ID", "브랜드코드"])["구매가격"].sum().reset_index()
brand_purchase = tmp.loc[tmp.groupby("ID")["구매가격"].idxmax()]
brand_purchase.columns = ["ID", "최다브랜드코드", "최다브랜드금액"]
test_ft = test_ft.merge(brand_purchase, how="left", on="ID")
test_ft
```

## 결측치 처리

```python
train_ft.isnull().sum()
```

```python
test_ft.isnull().sum()
```

```python
train_ft = train_ft.fillna(0)
test_ft = test_ft.fillna(0)

train_ft.isnull().sum().sum(), test_ft.isnull().sum().sum()
```

## Feature Encoding

- ID 컬럼 제거

```python
train_ft = train_ft.drop(columns=["ID", "Unnamed: 0"])
test_ft = test_ft.drop(columns=["ID", "Unnamed: 0"])
```

```python
cols = train_ft.select_dtypes("object").columns.tolist()
cols
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown="ignore")
```

```python
tmp = pd.DataFrame(
    enc.fit_transform(train_ft[cols]).toarray(),
    columns =  enc.get_feature_names_out()
)

train_ft = pd.concat([train_ft,tmp],axis=1).drop(columns=cols)
train_ft.head()
```

```python
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)
test_ft = pd.concat([test_ft,tmp],axis=1).drop(columns=cols)
test_ft.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_ft[train_ft.columns] = scaler.fit_transform(train_ft)
train_ft.head()
```

```python
test_ft[test_ft.columns] = scaler.transform(test_ft)
test_ft.head()
```

## 정답 데이터

# 데이터 분리(Data Split)
- 학습데이터와 검증데이터 분리

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_ft, target, test_size=0.2, random_state=42)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

# 모델 학습(Train Model)


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_valid)
pred[:5]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid,pred)
```
