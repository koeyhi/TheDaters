# Data leakage
- 정보 누설
- 머신러닝에서 Data leakage는 예측 시 사용할 수 없는 정보가 모델의 학습및 평가에 사용되는 것을 말한다.
- 모델 학습 과정에서 실제로는 알 수 없는 정보가 모델에 포함되어 예측 성능이 과대 평가될 수 있는 현상을 의미
- Data leakage 문제가 있으면 모델이 과대평가 되고 모델의 일반화 가능성이 보장되지 않는다.

# Data leakage 발생 예시
-  실제 서비스 환경에서 예측할 때 사용할 수 없는 피처를 수집하여 학습 데이터와 테스트 데이터에서 적용하고 모델 학습 및 성능을 평가한 경우
    - ex) 영화 관객수를 예측하는 문제에서 스크린 수를 피처로 사용하는 경우
- 테스트 데이터의 정보를 모델의 학습에 사용한 경우

# 다른 샘플들의 통계치를 이용해서 피처를 생성하는 경우

- 범주4 별 수치1 에 대한 평균값을 피처로 추가

```python
tmp = train_customer.groupby("범주4")["수치1"].mean().reset_index().rename(columns={"수치1": "범주4_수치1_mean"})
tmp
```

```python
train_ft = pd.merge(train_ft, tmp, on="범주4", how="left")
train_ft.head()
```

## data leakage 발생 예시
- 테스트데이터를 집계할 때 다른 샘플을 이용하여 집계하여 피처를 추가할 경우 data leakage 발생

```python
tmp = test_customer.groupby("범주4")["수치1"].mean().reset_index().rename(columns={"수치1": "범주4_수치1_mean"})
pd.merge(test_ft, tmp, on="범주4", how="left")
```

## data leakage 해결 방안
- 학습데이터에서 집계한 데이터로 merge 해줘야 함

```python
tmp = train_customer.groupby("범주4")["수치1"].mean().reset_index().rename(columns={"수치1": "범주4_수치1_mean"})
test_ft = pd.merge(test_ft, tmp, on="범주4", how="left")
test_ft.head()
```

- 다음과 같은 경우는 data leakage 아님

```python
test_payment.groupby("고객ID")["카드한도"].mean()
```

# 결측치를 처리하는 경우

```python
agg_dict = {
    "카드현재잔액": [
        ("카드현재잔액_mean", "mean"),
        ("카드현재잔액_std", "std")
    ]
}
tmp = train_payment.groupby("고객ID").agg(agg_dict)
tmp.columns = tmp.columns.droplevel()
train_ft = pd.merge(train_ft, tmp.reset_index(), on="고객ID", how="left")
train_ft.head()
```

```python
tmp = test_payment.groupby("고객ID").agg(agg_dict)
tmp.columns = tmp.columns.droplevel()
test_ft = pd.merge(test_ft, tmp.reset_index(), on="고객ID", how="left")
test_ft.head()
```

```python
train_ft.isnull().sum()
```

```python
test_ft.isnull().sum()
```

## data leakage 발생 예시

```python
df = pd.concat([train_ft, test_ft]).reset_index(drop=True)
fill_value = df["카드현재잔액_std"].mean() # data leakage 발생
fill_value
```

```python
df["카드현재잔액_std"].fillna(fill_value)
```

## data leakage 해결 방안
- 테스트 데이터의 결측치를 처리할 때 특정 상수값으로 처리하거나 학습데이터에서 구한 통계치로 처리하면 된다.

```python
fill_value = train_ft["카드현재잔액_std"].mean()
```

```python
train_ft["카드현재잔액_std"] = train_ft["카드현재잔액_std"].fillna(0)
test_ft["카드현재잔액_std"] = test_ft["카드현재잔액_std"].fillna(0)
```
