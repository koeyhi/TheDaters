# 연산자

## 산술연산자

## 대입 연산자
- 값(객체)을 변수에 할당
    > `=`
- 복합 대입 연산자
    - 산술연산결과의 값을 변수에 할당
    > `+= , -= , ...`

```python
num = 19
num += 2 # num = num + 2
num
```

```python
num -= 2
print(num)
num *= 2
print(num)
num /= 2
print(num)
num //= 2
print(num)
num %= 5
print(num)
num **= 2
print(num)
```

## 비교 연산자
- 값(객체)을 비교하여 bool 객체를 반환
- `> , < , >=, <= , !=, ==`

```python
a = 5
b = 3
print(a < b)
print(a > b)
print(a <= b)
print(a >= b)
print(a != b)
print(a == b)
```

## 논리연산자
- 논리 자료형인 bool객체를 연산하여 bool 객체를 반환
- and, or, not
    - A and B : A와 B 모두 True 면 결과는 True, 아니면 False
    - A or B : A와 B 둘 중 하나만 True여도 결과는 True , 둘다  False면 False
    - not A : A가 True면 False가 되고, False면 True가 된다.


```python
print(1 < 3 and 3 < 1)
print(1 < 3 or 3 < 1)
print(not 0)
print(not 1)
```

## 포함 연산자
- 어떠한 값이 포함 되어있는지 여부에 따라 bool 객체를 반환


```python
a = 'Python'
print('th' in a)
print('Python' in a)
print('No' in a)
```

## is 와 is not 연산자
- 동일한 주소에 할당된 객체임을 비교하는 연산자로 bool 객체를 반환
- `is`: 동일한 주소에 할당된 객체이면 True, 아닐경우 False
- `is not`: 동일한 주소에 할당된 객체이면 False, 아닐경우 True

```python
a = 1001
b = 1001
print(a is b)
print(a == b)
print(a is not b)
```

1. 3명의 학생의 시험점수가 각 변수에 들어가 있다.
변수의 값들을 이용하여 평균을 구하고, 구한 평균을 이용하여 표준편차를 구하시오.

$$
\mu = \frac{1}{N}\sum_{i=1}^{N}x_i
$$
$$
\sigma^{2} = \frac{1}{N}\sum_{i=1}^{N}(x_i - \mu)^{2}
$$

```python
karns = 90
minsu = 87
gildong = 72
avg = 0
stdev = 0

avg = (karns + minsu + gildong) / 3
stdev = (((karns - avg) ** 2 + (minsu - avg) ** 2 + (gildong - avg) ** 2) / 3) ** 0.5

print(avg, stdev)
```
