# 주피터 노트북

## 셀의 모드
- 명령모드
    - a(위에 셀추가) , b(아래셀추가) , dd(셀삭제)
- 입력 모드
    - ctrl + enter( 셀실행후 현재 셀선택)
    - shift + enter( 셀실행후 아래 셀선택)
    - ctrl + / (주석처리)
    - shift + tab(내어쓰기)
    - ctrl + alt(다중 커서)

## 셀 종류
- 코드 cell, 마크다운 cell(텍스트 cell)
- 마크업 언어는 태그나 서식 등을 이용하여 문서나 데이터의 구조를 명시하는 언어에 일종
- 마크다운은 일반 텍스트 기반의 경량화된 마크업 언어
