## 문자열 포매팅(String Formatting)
- 문자열 내의 특정한 값을 변수를 이용해 삽입하는 방법
- 포맷 코드
    - `%s`: 문자열(str)
    - `%d`: 정수(int)
    - `%f`: 실수(float)

````python
name = "이관수"
age = 44
gpa = 4.5

text = "제 이름은 " + name + "입니다. 나이는 " + str(age) + "이고, 학점은 " + str(gpa) + "입니다."
text
````

### `%` 기호를 사용한 방식

````python
text = "제 이름은 %s입니다. 나이는 %d이고, 학점은 %.1f 입니다." % (name, age, gpa)
text
````

### format 메서드를 사용한 방식

````python
text = "제 이름은 {}입니다. 나이는 {}이고, 학점은 {}입니다."
text.format(name,age,gpa)
````

````python
text = "제 이름은 {}입니다. 나이는 {}이고, 학점은 {:.2f}입니다."
text.format(name,age,gpa)
````

### F-string 

````python
text = f"제 이름은 {name}입니다. 나이는 {age}이고, 학점은 {gpa}입니다."
text
````

````python
text = f"제 이름은 {name}입니다. 나이는 {age}이고, 학점은 {gpa:.2f}입니다."
text
````
