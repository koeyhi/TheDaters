# PIP 란?
- Package installer for Python
- Python 패키지를 설치하고 관리하는 패키지 매니저
- Python 3.4 이후 버전은 설치되어있음
- pip을 이용하면 패키지를 쉽게 설치
- pip로 설치할 수 패키지들의 목록은 다음의 URL에서 찾아 볼 수 있음
- https://pypi.org/
- pip 명령어 실행시 주피터 노트북 환경에서는 `%` or `!` 앞에 붙여야 한다.


- 현재 설치된 패키지 리스트와 버전을 출력
```bash
pip list
```

- 패키지 설치
```bash
pip install <패키지이름>
```

- 패키지 삭제
```bash
pip uninstall <패키지이름>
```

- 설치되어있는 패키지 기록해놓기
```bash
pip freeze > requirements.txt
```

- 기록해 놓은 파일을 이용하여 개발환경 복구하기
```bash
pip install -r requirements.txt
```

# 파이썬의 가상환경이란?
- 독립적인 파이썬 실행 환경
- 파이썬 가상환경의 주요 목적은 파이썬 프로젝트를 위한 격리된 환경을 만드는 것이다.
- 각각의 가상환경은 다른 가상환경에 영향을 미치치 않음
    - Python 3.4 이후 버전은 설치 없이 기본적으로 지원
    - 사용자가 지정한 폴더에 파이썬및 기본패키지들이 설치

- 파이썬 개발 환경 가상화 필요성
    - 한 대의 컴퓨터에 여러 버전의 파이썬 패키지를 사용할 경우
    - 한 대의 컴퓨터에 여러 사용자가 파이썬 개발을 할 경우

## 가상환경 명령어
- 가상환경 생성하기
```bash
python -m venv <가상환경이름>
```

- 가상환경으로 진입하기(Windows)
```
{가상환경이름}\Scripts\activate
```

- 가상환경으로 진입하기(Linux)
```bash
source {가상환경이름}/bin/activate
```
- 가상환경에서 빠져나오기
```bash
deactivate
```
