


# 파일 입출력

- open 함수: 파일을 열수 있는 함수, 파일 객체를 반환한다.
- 파일열기 모드는 대표적으로 다음과 같다.
    - w: 쓰기 모드
    - a: 추가 모드
    - r: 읽기 모드
    
    

- open 함수로 열면 파일 객체가 반환 되는데 파일에 대한 처리가 종료되면 항상 close 메소드를 이용해서 닫아줘야한다.

````python
DATA_PATH = "data/"
````

````python
import os
if not os.path.isdir(f"{DATA_PATH}"):
    os.mkdir(f"{DATA_PATH}")
````

- 쓰기모드

````python
f = open(f"{DATA_PATH}memo.txt", "w")
print(f)
f.close()
````

````python
f = open(f"{DATA_PATH}memo.txt", "w", encoding="utf-8")

for i in range(1,11):
    f.write(f"{i} 번째 줄입니다. \n")

f.close()
````

- 추가모드

````python
f = open(f"{DATA_PATH}memo.txt", "a", encoding="utf-8")

f.write(f"11 번째 줄입니다.")

f.close()
````

- 읽기 모드

````python
f = open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8")

data = f.read() # 파일 전체 내용을 읽어서 문자열로 반환

f.close()
````

````python
print(data)
````

````python
f = open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8")

while True:
    data = f.readline()
    print(data, end="")
    if not data:
        break
f.close()
````

````python
f = open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8")

for i in f.readlines():
    
    print(i.strip())

f.close()
````

````python
f = open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8")

data = f.readlines()
    
f.close()

data
````

- with 문
    - 자원을 획득해서 사용 후 반납할 때 자동으로 반납해줌.

````python
with open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8") as f:
    data = f.read()
data
````

````python
f = open(f"{DATA_PATH}memo.txt", "r", encoding="utf-8")
````

````python
for i in f: 
    print(i)
````

````python
f.close()
````

# JSON(JavaScript Object Notation)
- 자바스크립트에서 사용되는 객체 표현식
- 데이터를 저장하거나 전송할때 많이 사용
- 서버와 클라이언트간의 데이터 전송과정에서 많이 사용
- 파이썬의 딕셔너리와 매우 유사

````python
import json
````

````python
data = {
    "karns" : {
        "age": 44,
        "phone" : "010-0000-0000",
        "city" : "seoul"
    },
    "관수" : {
        "age": 30,
        "phone": "010-1111-1111"
    }
}
````

- json 파일 저장하기

````python
with open(f"{DATA_PATH}info.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4) 
````

- json 파일 불러오기

````python
with open(f"{DATA_PATH}info.json", "r", encoding="utf-8") as f:
    data = json.load(f)  # 딕셔너리 형태로 반환

type(data)
````

### YAML (Yet Another Markup Language)
- 사람이 쉽게 읽을 수 있는 데이터 양식
- 데이터를 저장하거나 전송할때 많이 사용

```yaml
karns:
    age: 44
    phone: 010-0000-0000
    city: seoul
관수:
    age: 30
    phone: 010-1111-1111

리스트:
    - 320
    - 234
    - "12345"  # 숫자를 문자열로 저장하고 싶을경우 따옴표로 감싸줘야함.

```

# pickle 모듈
- 우리가 사용하는 다양한 객체들을 파일로 저장하고 싶다면?
- 객체의 구조를 직렬화해서 파일로 저장한다.
- 보통 딕셔너리, 리스트 등을 저장하고 불러올때 사용한다.


````python
import pickle
````

````python
data = [1,2,3,4,[10,20],"hi"]
````

````python
with open(f"{DATA_PATH}data.pkl", "wb") as f:
    pickle.dump(data, f)
````

````python
with open(f"{DATA_PATH}data.pkl", "rb") as f:
    lst = pickle.load(f)
lst
````

````python
class MyClass:
    def __init__(self):
        self.my_name = "My Class"
````

````python
mc = MyClass()
````

````python
with open(f"{DATA_PATH}myclass.pkl", "wb") as f:
    pickle.dump(mc, f)
````

````python
with open(f"{DATA_PATH}myclass.pkl", "rb") as f:
    mc_obj = pickle.load(f)
mc_obj
````

````python
def do_func():
    return "hi"
````

````python
with open(f"{DATA_PATH}dofunc.pkl", "wb") as f:
    pickle.dump(do_func, f)
````

````python
with open(f"{DATA_PATH}dofunc.pkl", "rb") as f:
    fc_obj = pickle.load(f)
fc_obj
````
