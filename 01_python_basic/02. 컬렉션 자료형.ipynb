### 슬라이싱(Slicing)
- 인덱스를 사용해 범위를 지정해서 요소들을 가져오는 방법
- 시퀀스 자료형에서 특정 범위의 요소들을 선택해서 가져올때 사용
- `[start : end : step]`
    - `start` ~ `end-1` 부분을 `step`만큼 건너 뛰면서 선택해서 가져온다.
    - `start` 와 `end` 값은 자료에서 가져올 범위를 지정하는 번호인데 `end` 번호는 포함하지 않고 가져온다.

````python
lst = ["a", "b", "c", "d", "e", "f", "g"]
lst
````

````python
lst[0:3] # 0~2 번에 인덱스에 해당하는 요소들이 반환
````

````python
lst[2:5] # 2~4 번에 인덱스에 해당하는 요소들이 반환
````

````python
# start 와 end를 생략할 경우 자료에서 처음부터 끝까지 가져온다.
lst[:] ## lst[0:len(lst)] 
````

````python
a = lst[:] # 다른 변수에 복사하고 싶다??
a
````

````python
lst[1:] # end 값을 생략할 경우 지정한 start 값부터 끝까지 가져온다.
````

````python
lst[:-1] # start 값을 생략할 경우 처음부터 지정한 end 값을 포함하하지 않고 end 값 바로 전 순번까지 가져온다. 
````

### step

## 패킹(packing)과 언패킹(unpacking)
- 패킹
    - 여러 값(객체)들을 하나의 변수에 담기
- 언패킹
    - 하나의 변수에 담긴 여러값들을 여러 변수에 풀기

````python
num = [1,2,3] # 패킹
````

````python
a, b, c = num # 언패킹

print(a)
print(b)
print(c)
````

# 숙제1.
- salary 변수에 직원들의 연봉을 담은 리스트가 담겨 있다.
- 가장 높은 연봉을 제외해서 평균 연봉을 구하시오.

````python
salary = [3500,4000,4500,6000,3600,8000,15000,3500,4000,4500,6000,3600] # 만단위
````

````python
tmp = sorted(salary)[:-1]
sum(tmp) / len(tmp)
````
